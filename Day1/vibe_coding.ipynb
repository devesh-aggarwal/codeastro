## What is vibe coding
 * Vibe coding is using artificial intelligence (AI; in particular, generative AI such as those based on large language models) to generate source code automatically by prompting AI tools. 
     * The prompts can be not very specific, only providing the general "vibe" of what the user is looking for, and are written in a natural language (such as English). 
     * Some of these tools are described as "AI agents" or "agentic AI", which typically means that the vibe coding tool is able to go beyond just generating code from a prompt and also automate, plan, and/or oversee the coding themselves as well. 

### AI offerings

There are many AI services to use for vibe coding (Google Gemini, Anthropic Claude, ChatGPT, Microsoft Copilot, etc.). Some have free tiers, and some are paid. In the context of software development, some of the main differences are:

 * Pricing: some services (Google, Microsoft) have free tiers, others (Anthropic's Claude Code service in particular) do not. For paid models, some are flat fees while others charge based on usage. 
 * Model performance: the more advanced models that perform better generally are more expensive to run and are slower. Often the free tiers only give access to the most basic models, while "frontier" models can sometimes do more sophiticated tasks. 
 * Context awareness: how much information on your project do they know.
     * the AI running in the chat window of your browser does not know any context about the code you are writing. 
     * Services integrated into IDEs or the command line (Claude Code's VS Code integrate) can read your folders and files to understand the context of the code and how it interacts with other pieces of the software. 
     * Google Gemini's Colab Integration can only access the file that's currently open (I think). 
 * Allowed capabilties: similar to the above, but more focused on what kind of permissions does it have.
     * Some can only output chat prompts
     * Some can directly modify your code as well
     * Some can run arbitrary command line commands (which can be very powerful for automation, but also comes with a lot of risks). For example, because Claude Code can execute arbitrary code (with your permission), it can follow links or look up information by itself (e.g., you can tell it to "fix issue 25 on Github" and it will go online and read the issue itself, whereas Gemini's Colab Integration cannot.).

## General Recommendations in Approaching Vibe Coding

### Philosophical Recommendations
There is some psychological tendency to want to cut corners when using AI tools. But we should not cut corners to maintain high quality science. Below are some recommendations for how to maintain high-quality science while vibe coding.

 * You need to understand the code you write, even if you prompted AI to write it
     * For open-source software packages, don't commit code you don't understand!
         * This is different from When you're testing, getting familiar with a package, or just trying to understand something.
     * This is especially important in science. Our field and our individual reputation is built upon the understanding of how things work. Science requires understanding of how, not just whether the final answer looks correct. 
     * Understanding takes time and effort, and there is not much to can do to get around it. 
 * Review every line of code generated by an AI tool
     * "You need to understand not just your prompts, but you need to understand the end result too, because that's the only way you can maintain it long term." ~Linus Torvald, creator of Linux.
     * If you are contributing to a repository, review your changes before you submit a pull request (a pull request submits a chunk of new changes for the maintainer to consider).
     * If you are reviewing code, review every line yourself. Running an AI tool on it is not the same as reviewing it yourself. 
 * You are responsible for your code, even if it was made with AI
     * Science is a human endeavor and we use tools to accomplish the task. AI can be on one of those tools, but it is still the scientist that produces good science.
     * Do not hide behind or blame the AI tool. The onus, like it always has been, is on you.
* The physical act of writing code is just one part of the software development process, which AI can speed up. However, it does not speed up other aspects of software development like understanding and internalizing the code.
     * Because of this, vibe coding does not speed things up by orders of magnitude, but it can speed up software development noticably. 
* Generative AI is particular strong at programming compared to other tasks. Thus, not everything is solved by AI. However, many programming tasks can be solved by AI nowadays.
* At Code/Astro, we don't endorse tool shaming. We should not make someone who either does or does not use AI tools feel bad. 



### Vibe Coding Tips

 * Vibe coding tools that can integrate into an IDE generally are more powerful and are more capable. They do carry risk of accessing potentially your entire machine though, so be careful with them.
 * LLMs are better at smaller and more focused tasks. Vibe coding tools like Claude Code try to break down complicated problems down to simpler problems to solve them. If you can already do that yourself, it'll make vibe coding more effective.
 * LLMs perform worse are context increases due to the fundamental nature of how they are built. Compactifying the chat history or starting a new chat can lead to better performance. This is because your entire chat history gets put in as input into your next message, increasing the size of how many tokens are passed in as input, and makes it harder on the "attention" of the LLM. 
 * LLMs are better at pure "programming" problems, and less good at knowing how to properly handle external state (e.g., the data that needs to be processed here generally has only a few artifacts, or the signal to noise level is always kind of low). If you can interpret the external state and turn it into a programming decision, that makes vibe coding easier (because there are only a few artifacts per image, a for loop is good enough to identify them; we should avoid trusting the maximum value when the signal to noise level is low). 
 * As a collorary, because LLMs are fairly effective at pure "programming" problems, they are actually be quite useful for a single person maintaining a open-source code repository. I've been able to finally fix a bunch of minor, but annoying bugs in packages I maintain thanks to vibe coding.
 * LLMs like creating very condensed (e.g., one-line) solutions to programming problems, which often are hard to read. You can discourage them from doing so or ask it to rewrite changes. Remember, we want to understand each line of code!
 * Some vibe coding tools like Claude Code allow you to setup a instruction file (e.g., CLAUDE.md) to provide guidance on how it should operate. Note that these are suggestions, and I've found the LLMs to break them, so putting a rule there is not set in stone! 
 * You can prompt LLMs to first plan changes, output you a list of planned changes, edit the list of planned changes to your specifications, and then prompt it to follow the written out list of changes. This is what Claude Code does under the hood to some extent, but it is sometimes more effective if you force it to do it explicitly. It also offers you a chance to inspect the plan at a high level before it makes lower level changes. 
     * Even better, if you can provide a list of requirements or a design document, this is greatly helpful. For example, if you can write out a list of test cases or use cases that need to succeed.
